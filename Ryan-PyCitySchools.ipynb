```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data = "schools_complete.csv"
student_data = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete.head(5)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schools=school_data_complete.set_index('school_name')
each_school=schools['budget'].unique()
each_school
total_budget=each_school.sum()
total_budget
total_schools=len(each_school)
total_students=len(school_data_complete['Student ID'])
total_schools,total_students,total_budget
avg_math=school_data_complete['math_score'].mean()
avg_read=school_data_complete['reading_score'].mean()
avg_math,avg_read
math_pass=school_data_complete.loc[:,'math_score']>=70
math_pass=math_pass.sum()
read_pass=school_data_complete.loc[:,'reading_score']>=70
read_pass=read_pass.sum()
math_percent=(math_pass/total_students)
math_percent
read_percent=(read_pass/total_students)
read_percent
overall_pass=school_data_complete[(school_data_complete['math_score']>=70)&(school_data_complete['reading_score']>=70)].count()['student_name']
overall_pass=(overall_pass/total_students)
math_percent,read_percent,overall_pass

format_dict={'Total Budget':'${0:,.0f}',
             'Math Passing Percentage':'{:.2%}',
             'Reading Passing Percentage':'{:.2%}',
             'Overall Passing Percentage':'{:.2%}'
        }
district_summary=pd.DataFrame({
    'Total Schools':[total_schools],
    'Total Students':[total_students],
    'Total Budget':[total_budget],
    'Average Math Score':[avg_math],
    'Average Reading Score':[avg_read],
    'Math Passing Percentage':[math_percent],
    'Reading Passing Percentage':[read_percent],
    'Overall Passing Percentage':[overall_pass]
})
district_summary.style.format(format_dict).hide_index()    


#len(total_schools),total_students,total_budget,avg_math,avg_read
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading

* Create a dataframe to hold the above results

```python
# #rename the complete data set for ease
full_df=school_data_complete
# #school distract summary doesn't need every series. Remove the ones not needed
# #You need to calculate budget per student (budget/size), math avg and percent pass, read av and percent pass
# #Plus the overall pass of each school
school_district=full_df[['student_name','school_name','reading_score','math_score','type','size','budget']]
school_district['Passed Math']=school_district['math_score']>= 70
school_district['Passed Reading']=school_district['reading_score']>=70
school_district['Passed Both']=(school_district['math_score']>=70)&(school_district['reading_score']>=70)
passed_both=school_district[school_district['Passed Both']== True]
passing_math=school_district[school_district['math_score']>= 70]
passing_read=school_district[school_district['reading_score']>=70]
school_count=school_district['school_name'].value_counts()

                    

per_math=passing_math.groupby('school_name').count()['student_name']/school_count*100
per_read=passing_read.groupby('school_name').count()['student_name']/school_count*100
per_both=passed_both.groupby('school_name').count()['student_name']/school_count*100  
                             


school_district=school_district.groupby('school_name')
school_type=school_district['type'].unique()
school_type=[x[0] for x in school_type]
total_students=school_district['size'].unique()
school_budget=school_district['budget'].unique()
per_student_budget=(school_budget/total_students)
per_student_budget=[x[0] for x in per_student_budget]
school_budget=[x[0] for x in school_budget]
total_students=[x[0] for x in total_students]
# avg_math=school_district['math_score'].mean()
# avg_read=school_district['reading_score'].mean()



# #Take your mathpassers grouped by school name then count student_name 
# #divide by the value counts of school_name times 100
format_dict={
    "School's Budget": '${0:,.0f}',
    "Per Student Budget": '${0:,.0f}'
}
school_district=pd.DataFrame({
    "School Type": school_type,
    "Students Per School": total_students,
    "School's Budget": school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": per_math,
    "% Passing Reading":per_read,
    "% Passing Overall": per_both
})


school_district.style.format(format_dict)
                             
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#use full_df and average Math score series broken down by grade. Grade is also a series
#parse down full_df to the needed Series
math_grade=full_df.loc[:,['school_name','grade','math_score']]
math_grade
#conditional statement to pull out each Grade, set to variable
nine_graders=math_grade[(math_grade['grade']=='9th')]
ten_graders=math_grade[(math_grade['grade']=='10th')]
eleven_graders=math_grade[(math_grade['grade']=='11th')]
twelve_graders=math_grade[(math_grade)['grade']=='12th']
#use groupby.mean on each grade variable
nine_scores=nine_graders.groupby(['school_name']).mean()['math_score']
ten_scores=ten_graders.groupby(['school_name']).mean()['math_score']
eleven_scores=eleven_graders.groupby(['school_name']).mean()['math_score']
twelve_scores=twelve_graders.groupby(['school_name']).mean()['math_score']

math_scores=pd.DataFrame({
    "9th":nine_scores,
    "10th":ten_scores,
    "11th":eleven_scores,
    "12th":twelve_scores
})
math_scores
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#start with full_df and parse down to school_name,grade, and reading_score
read_grade=full_df.loc[:,['school_name','grade','reading_score']]
read_grade
#pull each grade out with a conditional statement
nine=read_grade[(read_grade['grade']=='9th')]
ten=read_grade[(read_grade['grade']=='10th')]
eleven=read_grade[(read_grade['grade']=='11th')]
twelve=read_grade[(read_grade['grade']=='12th')] 
#use groupby['school_name'].mean()['reading_score'] set to variable
nine_scores=nine.groupby(['school_name']).mean()['reading_score']
ten_scores=ten.groupby(['school_name']).mean()['reading_score']
eleven_scores=eleven.groupby(['school_name']).mean()['reading_score']
twelve_scores=twelve.groupby(['school_name']).mean()['reading_score']

read_grades=pd.DataFrame({
    '9th': nine_scores,
    '10th': ten_scores,
    '11th': eleven_scores,
    '12th': twelve_scores
})

read_grades
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#using bins of our per student budget, cull down school_district with our needed series(perstudent,math-avg,read-avg, and 
# all of the percents)
spending_df=school_district.loc[:,['Per Student Budget','Average Math Score','Average Reading Score',
                                  '% Passing Math','% Passing Reading','% Passing Overall']]
#create your bins values and labels
bins=[0,585,630,645,675]
bin_names=['<$585','$585-629','$630-644','$645-675']
spending_df['Spending Ranges(Per Student)']=pd.cut(spending_df['Per Student Budget'], bins, labels=bin_names)
#get scores for 5 categories
avg_math=spending_df.groupby(['Spending Ranges(Per Student)']).mean()['Average Math Score']
avg_read=spending_df.groupby(['Spending Ranges(Per Student)']).mean()['Average Reading Score']
per_math=spending_df.groupby(['Spending Ranges(Per Student)']).mean()['% Passing Math']
per_read=spending_df.groupby(['Spending Ranges(Per Student)']).mean()['% Passing Reading']
per_all=(avg_math + avg_read)/2

school_spending_df = pd.DataFrame({
    'Average Math Score': avg_math,
    'Average Reading Score': avg_read,
    '% Passing Math': per_math,
    '% Passing Reading': per_read,
    '% Passing Overall': per_all
})
school_spending_df

```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Same as above, just using school size to set bins. 
size_df=school_district.loc[:,['Students Per School','Average Math Score','Average Reading Score',
                                  '% Passing Math','% Passing Reading','% Passing Overall']]
#create bins for school size
bin=[0,1000,2000,5000]
bin_sizes=['Small(<1000)', 'Medium(1000-2000)', 'Large(2000-5000)']
size_df['School Size']=pd.cut(size_df['Students Per School'], bin, labels=bin_sizes)
#groupby scores and averages
avg_math=spending_df.groupby(['Spending Ranges(Per Student)']).mean()['Average Math Score']

avg_math=size_df.groupby(['School Size']).mean()['Average Math Score']
avg_read=size_df.groupby(['School Size']).mean()['Average Reading Score']
per_math=size_df.groupby(['School Size']).mean()['% Passing Math']
per_read=size_df.groupby(['School Size']).mean()['% Passing Reading']
per_all=(avg_math+avg_read)/2

school_size_df=pd.DataFrame({
    'Average Math Score': avg_math,
    'Average Reading Score': avg_read,
    '% Passing Math': per_math,
    '% Passing Reading': per_read,
    '% Passing Overall': per_all
})
school_size_df

```
